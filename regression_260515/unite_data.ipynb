```python
import pandas as pd

# CSV 파일 불러오기
df1 = pd.read_csv("simulation_results1.csv")
df2 = pd.read_csv("simulation_results2.csv")
df3 = pd.read_csv("simulation_results3.csv")
df4 = pd.read_csv("simulation_results4.csv")
df5 = pd.read_csv("simulation_results5.csv")

# 두 데이터프레임 합치기
df_combined = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# 합친 데이터프레임 저장
df_combined.to_csv("simulation_results.csv", index=False)

df_combined
```
